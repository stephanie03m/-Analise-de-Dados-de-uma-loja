```python
import pandas as pd
from pandas import merge, DataFrame
```

```python
online_orders = pd.read_json('online_orders.json', lines= True)
OfflineSales = pd.read_json('offline_sales.json', lines= True)
```

```python
#Número de cliene que copram na loja física
totalloja = OfflineSales.customer_id.nunique()
totalloja
```

```python
#Número de cliene que copram online
totalonline = online_orders.customer_id.nunique()
totalonline
```

```python
online_orders.dropna(inplace=True)
OfflineSales.dropna(inplace=True)
```

```python
online_orders
```

```python
loja_online = pd.merge(OfflineSales,online_orders,on='customer_id', how = 'inner')
loja_online
```
