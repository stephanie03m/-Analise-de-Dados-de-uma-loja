```python
import pandas as pd
from pandas import merge
```

```python
OfflineSales = pd.read_json('offline_sales.json', lines= True)
online_pageviews = pd.read_json('online_pageviews.json', lines= True)
```

```python
online_pageviews
```

```python
online_pageviews.dropna(inplace=True)
```

```python
loja_online = pd.merge(OfflineSales,online_pageviews,on='customer_id', how = 'inner')
loja_online
```

```python
clientes = loja_online.groupby('state')['customer_id'].nunique()
clientes
```
